```python
import pandas as pd

data = pd.read_csv('stations.csv')
data.head()
```

```python
class Station:
    def __init__(self, name, line):
        self.name = name 
        self.line = line 
        self.connections = []

    def __str__(self):
        return self.name + " | Line: " + self.line

stations = []

for row in data.values:
    lines = row[2].split('/')
    for line in lines:
        station = Station(row[1], line)
        stations.append(station)

for s in stations:
    print(s)
```
